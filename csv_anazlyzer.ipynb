```python
with open("students.csv", "r", encoding= "utf-8-sig") as file:
    lines = file.readlines()
    total = 0
    count = 0
    for line in lines[1:]:
        data = line.strip().split(",")
        marks = int(data[3])
        total = total+marks
        count = count+1
    average = total/count
print(average)

```

```python
#topper
with open("students.csv", "r", encoding = "utf-8-sig") as file:
    lines = file.readlines()
    highest_marks=0
    topper = ""
    for line in lines[1:]:
        data = line.strip().split(",")
        marks = int(data[3])
        if marks>highest_marks:
            highest_marks=marks
            topper=data[0]
    print("Topper:", topper)
    print("Marks:", highest_marks)

```
